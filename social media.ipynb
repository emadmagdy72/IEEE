```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
dic = dict()
names = []
authors = []
publishers = []
years = []

for i in range(1,11):
    url = f'https://eg1lib.org/s/social%20media?page={i}'
    r = requests.get(url)
    html_doc = r.text
    soup = BeautifulSoup(html_doc)
    containers = soup.find_all('div',{'class':'resItemBox resItemBoxBooks exactMatch'})
    for j in range(len(containers)):
        names.append(containers[j].div.h3.a.text)
        authors.append(containers[j].find_all('div',{'class':'authors'})[0].text)
        publishers.append(containers[j].div.tr.tr.div.text)
        years.append(containers[j].find_all('div',{'class':"property_value"})[0].text)
```

```python
dic = {'Name':names, 'Publisher':publishers, 'Author':authors,'Year':years}
```

```python
df = pd.DataFrame(dic)
df
```
